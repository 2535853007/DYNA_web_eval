```python
import os
import json

def extract_website_defects(base_path, model_name):
    """
    遍历目录，提取 judge_false.json 中的 website_defect_reason。
    """
    results = {}
    
    if not os.path.exists(base_path):
        print(f"❌ 路径不存在: {base_path}")
        return

    # 遍历任务类型文件夹 (例如: Accommodation Booking__lovable)
    for subdomain in os.scandir(base_path):
        if not subdomain.is_dir():
            continue
            
        # 遍历具体的任务文件夹 (例如: Find the taxes and fees...)
        for task_dir in os.scandir(subdomain.path):
            if not task_dir.is_dir():
                continue
            
            # 拼接 judge_false.json 的路径
            false_json_path = os.path.join(task_dir.path, "judge_false.json")
            
            if os.path.exists(false_json_path):
                try:
                    with open(false_json_path, 'r', encoding='utf-8') as f:
                        data = json.load(f)
                        
                        # 获取 website_defect_reason 内容
                        # 如果字段不存在，则返回 None
                        reason = data.get("website_defect_reason")
                        
                        # 只有当理由不为空（且存在）时才记录
                        if reason is not None:
                            # 文件夹名称作为 Key，理由作为 Value
                            results[task_dir.name] = reason
                            
                except Exception as e:
                    print(f"读取文件出错 {false_json_path}: {e}")

    # 确保保存目录存在
    save_dir = 
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
        
    json_file_path = os.path.join(save_dir, "defect_tasks.json")

    # 写入最终的 json 文件
    try:
        with open(json_file_path, 'w', encoding='utf-8') as f:
            json.dump(results, f, ensure_ascii=False, indent=4)
        print(f"✅ 任务提取完成，保存至: {json_file_path}")
        print(f"📊 共提取了 {len(results)} 条缺陷记录。")
    except Exception as e:
        print(f"保存文件失败: {e}")

if __name__ == "__main__":
    # 请根据实际情况修改 model_name
    model_name = "bolt2" 
    target_path = 
    
    extract_website_defects(target_path, model_name)
```

```python
import os
import json

def generate_defect_reports(base_path, model_name):
    """
    为每个任务类型文件夹生成一个包含所有 website_defect_reason 的 .md 文件。
    """
    # 设定保存路径
    save_dir =
    
    if not os.path.exists(save_dir):
        os.makedirs(save_dir, exist_ok=True)
        print(f"📂 已创建目录: {save_dir}")

    if not os.path.exists(base_path):
        print(f"❌ 路径不存在: {base_path}")
        return

    # 1. 遍历任务类型文件夹 (例如: Accommodation Booking__lovable)
    for subdomain in os.scandir(base_path):
        if not subdomain.is_dir():
            continue
        
        type_name = subdomain.name
        reasons_list = []
        
        # 2. 遍历该类型下的所有样本文件夹
        for task_dir in os.scandir(subdomain.path):
            if not task_dir.is_dir():
                continue
            
            false_json_path = os.path.join(task_dir.path, "judge_false.json")
            
            if os.path.exists(false_json_path):
                try:
                    with open(false_json_path, 'r', encoding='utf-8') as f:
                        data = json.load(f)
                        reason = data.get("website_defect_reason")
                        
                        # 只收集有内容的理由，且过滤掉空值
                        if reason and str(reason).strip():
                            reasons_list.append(str(reason).strip())
                except Exception as e:
                    print(f"读取文件出错 {false_json_path}: {e}")

        # 3. 如果该类型下收集到了理由，则生成 md 文件
        if reasons_list:
            md_file_path = os.path.join(save_dir, f"{type_name}.md")
            
            with open(md_file_path, 'w', encoding='utf-8') as md_file:
                # 按照你的格式要求：数字编号. 理由内容
                for i, r in enumerate(reasons_list, 1):
                    # 如果理由末尾没有句号，可以根据需要自行补充
                    content = r if r.endswith(('.', '!', '?')) else r + '.'
                    md_file.write(f"{i}. {content}\n")
            
            print(f"✅ 已生成: {type_name}.md (包含 {len(reasons_list)} 条记录)")
        else:
            print(f"⚠️ 跳过 {type_name}: 未发现有效的 website_defect_reason")

if __name__ == "__main__":
    # 配置参数
 
    target_path =
    
    generate_defect_reports(target_path, model_name)
```
